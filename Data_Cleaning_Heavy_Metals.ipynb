_Remove Columns With More than 15 % of missing values__

```python
# Step 1: Calculate the percentage of missing values in each column
missing_percentage = (df_metals.isnull().sum() / len(df_metals)) * 100

# Step 2: Identify the columns where the percentage of missing values exceeds 15%
columns_to_remove = missing_percentage[missing_percentage > 15].index

# Step 3: Remove those columns from the DataFrame
df_metals.drop(columns=columns_to_remove, inplace=True)
```
