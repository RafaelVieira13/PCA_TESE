```python
import os
import requests
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time
```

# Get Heavy Metals Data

Let´s Scrape the all the PDF´S files from the following website: https://www.ipma.pt/pt/bivalves/metais/

```python
# Set Options to Download PDF files (will not be used for direct download)
options = webdriver.ChromeOptions()
options.add_experimental_option('prefs', {
    'download.default_directory': 'C:/Users/Administrador/Article/pdfs_heavy_metals',
    'download.prompt_for_download': False,
    'plugins.always_open_pdf_externally': True
})

# Set up the Selenium Driver (not used for downloading PDFs)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
wait = WebDriverWait(driver, 10)

# Let's open the website
driver.get('https://www.ipma.pt/pt/bivalves/metais/')

# Find all the elements with 'a' tags
all_a_tags = driver.find_elements(By.XPATH, '//a')

# Let's find all the pdf's files with 'a' tags and in the attribute 'href'
for a in all_a_tags:
    href = a.get_attribute('href')
    if href and '.pdf' in href:
        # Download the PDF file using requests
        response = requests.get(href)
        if response.status_code == 200:
            file_name = href.split("/")[-1]  # Extract the file name from the URL
            file_path = os.path.join('C:/Users/Administrador/Article/pdfs_heavy_metals', file_name)
            with open(file_path, "wb") as file:
                file.write(response.content)
            print(f"File '{file_name}' downloaded and saved successfully.")
        else:
            print(f"Failed to download file from {href}. Status code: {response.status_code}")

# Close the Selenium WebDriver (not used for downloading PDFs)
driver.quit()
```
