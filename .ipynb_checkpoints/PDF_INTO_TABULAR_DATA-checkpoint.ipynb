```python
import os
import pandas as pd
import tabula
warnings.filterwarnings("ignore")
```

# Convert Heavy Metals PDF´S Into Pandas Dataframe

```python
# Replace 'folder_path' with the path to your folder containing the PDF files.
folder_path = 'C:/Users/Administrador/Article/pdfs_heavy_metals/'

# Get a list of all PDF files in the folder
pdf_files = [file for file in os.listdir(folder_path) if file.endswith('.pdf')]

# Initialize an empty list to store the DataFrames
dfs_list = []

# Loop through each PDF file and read its contents into a DataFrame
for pdf_file in pdf_files:
    pdf_file_path = os.path.join(folder_path, pdf_file)
    dfs = tabula.read_pdf(pdf_file_path, pages='all', java_options='-Xmx1040m',guess=True) # specify the amount of memory that is available to the Java Virtual Machine (JVM). By increasing the amount of memory, you can allow the JVM to process more complex PDF files and extract more accurate data.
    dfs_list.extend(dfs)  # Append the DataFrames to the list

# Concatenate all DataFrames into a single DataFrame (optional)
combined_df = pd.concat(dfs_list, ignore_index=True)

# Print the number of PDF files processed and the total number of DataFrames obtained
print(f"Number of PDF files processed: {len(pdf_files)}")
print(f"Total number of DataFrames obtained: {len(dfs_list)}")
```

```python
combined_df
```

```python
combined_df.to_excel('combined_df.xlsx')
```
