#### Assumptions
    - 1. Independence: The observations should be independent of each other;
    - 2. Normality: The residuals should follow a normal distribution;
    - 3. Homogeneity of Variance: The variances of the groups being compared should be equal

__Normality__

_[Pb] by "Espécie"_

1.Histogram for each Specie, Shapiro-Wilk Test and QQ-Plot

```python
# Histogram
for species, species_data in df.groupby('Espécie'):
    sns.displot(species_data['Pb (mg/kg)'],kind='kde')
    plt.title(species)
```

```python
# Shapiro-Wilk Test
def shapiro_test(group_data):
    stat, p_value = stats.shapiro(group_data)
    alpha = 0.05
    print(f"Shapiro-Wilk Test for Normality - p-value: {p_value}")
    if p_value > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")
```

```python
groups = df.groupby('Espécie')['Pb (mg/kg)']

for group_name, group_data in groups:
    print(f"Group: {group_name}")
    shapiro_test(group_data)
```

```python
# QQ-plot 
for specie, species_data in df.groupby('Espécie'):
    stats.probplot(species_data['Pb (mg/kg)'], dist='norm',plot=plt,fit=True)
    plt.title(f'QQ-Plot for {specie}')
    plt.show()
```

__Homogeneity of Variance__

_"Espécie_

```python
# Levene Test
stat, p_value = stats.levene(*[group_data for _, group_data in groups])
alpha = 0.05
print(f"Levene's Test for Homogeneity of Variance - p-value: {p_value}")
if p_value > alpha:
    print("Homogeneity of Variance is met (fail to reject H0)")
else:
    print("Homogeneity of Variance is not met (reject H0)")
```

Conclusions: The Data doesn´t meet the anova ssumptions therefore let´s apply some transformations to modify the original in order to meet the anova assumptions.

__Perform Anova__

* In this case its one-way anova

```python
model = smf.ols('Q("Pb (mg/kg)") ~ Q("Espécie")', data=df).fit()
result = sm.stats.anova_lm(model)
print(result)
```

__Perfom Tukey Test__

```python
mc = MultiComparison(df['Pb (mg/kg)'], df['Espécie'])
result = mc.tukeyhsd()
print(result)
```

__Visualize the Residuals__

```python
# Visualize Residuals
residuals = model.resid
plt.scatter(df['Espécie'], residuals)
plt.xlabel('Espécie')
plt.xticks(fontsize=9,rotation=15)
plt.ylabel('Residuals')
plt.title('Residuals vs. Group')
plt.axhline(0, color='red', linestyle='--')
plt.show()
```

### Log-Transformations

```python
df.head()
```

```python
df['Pb_log'] = np.log(df['Pb (mg/kg)'])
df.head()
```

__Normality__

```python
# Histogram
for species, species_data in df.groupby('Espécie'):
    sns.displot(species_data['Pb_log'],kind='kde')
    plt.title(species)
```

```python
# Shapiro-Wilk Test
groups = df.groupby('Espécie')['Pb_log']

for group_name, group_data in groups:
    print(f"Group: {group_name}")
    shapiro_test(group_data)
```

```python
# QQ-plot
for specie, species_data in df.groupby('Espécie'):
    stats.probplot(species_data['Pb_log'], dist='norm',plot=plt,fit=True)
    plt.title(f'QQ-Plot for {specie}')
    plt.show()
```

__Homogeneity of Variance__

```python
# Levene test
stat, p_value = stats.levene(*[group_data for _, group_data in groups])
alpha = 0.05
print(f"Levene's Test for Homogeneity of Variance - p-value: {p_value}")
if p_value > alpha:
    print("Homogeneity of Variance is met (fail to reject H0)")
else:
    print("Homogeneity of Variance is not met (reject H0)")
```

__Perform Anova__

* In this case its one-way anova

```python
model = smf.ols('Q("Pb_log") ~ Q("Espécie")', data=df).fit()
result = sm.stats.anova_lm(model)
print(result)
```

__Perfom Tukey Test__

```python
mc = MultiComparison(df['Pb_log'], df['Espécie'])
result = mc.tukeyhsd()
print(result)
```

# Conclusions:

* The data doesn´t meet Anova Assumptions
* The One-Way Anova results with the original data shows that there is a significant difference between the [pb] mean among species. To understand the specific significant differences, the Tukey test was conducted, revealing significant differences in the [pb] mean among the following species:
    - Amêijoa-boa and Amêijoa-japonesa
    - Amêijoa-boa and Lambujinha
    - Amêijoa-japonesa and Lambujinha
    - Lambujinha and Mexilhão
    - Lambujinha and Ostra-japonesa
    - Lambujinha and Ostra-portuguesa
* To try to meet the ANOVA assumptions more closely, a log-transformation was performed, but it still did not fully meet the assumptions. Even after the transformation, another One-Way ANOVA showed significant differences in the [pb] mean among each species. Subsequently, the Tukey test was performed again, revealing significant differences between every species, except for the following pair:
    - Amêijoa-japonesa - Mexilhão
    
* As the ANOVA assumptions were not met, a non-parametric test will be conducted instead.
