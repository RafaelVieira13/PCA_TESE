# Merge Both Datasets

Let´s create a single dataset by merging the heavy metals and micro datasets based on the 'No Amostra' column

```python
df = pd.merge(df_metals,df_micro[['No Amostra','E.coli (NMP/100G)']], on='No Amostra',how='left')
df
```

__Chande the Data type of the columns 'Cd (mg/kg)', 'Hg (mg/kg)', 'Pb (mg/kg)' to float__

```python
columns = ['Cd (mg/kg)','Hg (mg/kg)','Pb (mg/kg)','E.coli (NMP/100G)']

for col in columns:
    print(df[col].value_counts())
```

As we can see before we have ',' instead of '.'

```python
# # Identify columns with commas as decimal separators
columns = ['Cd (mg/kg)','Hg (mg/kg)','Pb (mg/kg)','E.coli (NMP/100G)']

# Replace commas with periods, but keep missing values as they are
for col in columns:
    df[col] = df[col].apply(lambda x: x.replace(',', '.') if isinstance(x, str) else x)
                                                          # It checks if x is a string using the isinstance(x, str) condition. 
                                                          # If x is a string, it replaces any commas (',') with periods ('.') using x.replace(',', '.'). 
                                                          # If x is not a string (e.g., it is a numeric value or a missing value), it returns x unchanged.          
    
# Convert the data type to float
for col in columns:
    df[col] = pd.to_numeric(df[col])
```

# Quick Data Cleaning

Let´s clean the 'Data de Colheita' column and create a column for the day, month and year

```python
# Search for values with '-' ath the 'Data de colheita' column
df[df['Data de colheita'].str.contains('-')]
```

```python
# Replace every '-' by '/'
df['Data de colheita'] = df['Data de colheita'].str.replace('-','/')
```

```python
# Let´s create a column with the 'Day','Month' and 'Year' by split the 'Data de Colheita' using '/'
df[['Day','Month','Year']] = df['Data de colheita'].str.split('/',expand=True)
```

```python
df
```

__Let´s take a look into the 'Month' column__

* Only exists month between 1 (january) and 12 (december)
* Let´s convert every number to their respective month name
* Let´s analyse the month with a value above 12

```python
# Remove Spaces
df['Month'] = df['Month'].str.replace(' ','')
```

```python
df['Month'].value_counts()
```

```python
# Change some month names
df['Month'] = df['Month'].str.replace('ago','Aug')
df['Month'] = df['Month'].str.replace('Ago','Aug')
df['Month'] = df['Month'].str.replace('jul','Jul')
df['Month'] = df['Month'].str.replace('Set','Sept')
df['Month'] = df['Month'].str.replace('Abr','Apr')
df['Month'] = df['Month'].str.replace('Out','Oct')
```

```python
# Function to get the month names
def mes(mes):
    if mes == '1' or mes == '01':
        return 'Jan'
    elif mes == '2' or mes == '02':
        return 'Feb'
    elif mes == '3' or mes == '03':
        return 'Mar'
    elif mes == '4' or mes == '04':
        return 'Apr'
    elif mes == '5' or mes == '05':
        return 'May'
    elif mes == '6' or mes == '06':
        return 'Jun'
    elif mes == '7' or mes == '07':
        return 'Jul'
    elif mes == '8' or mes == '08':
        return 'Aug'
    elif mes == '9' or mes == '09':
        return 'Sept'
    elif mes == '10':
        return 'Oct'
    elif mes == '11':
        return 'Nov'
    elif mes == '12':
        return 'Dec'
    else:
        return mes  # Return the original value if it doesn't match any month
```

```python
df['Month'] = df['Month'].apply(mes)
```
